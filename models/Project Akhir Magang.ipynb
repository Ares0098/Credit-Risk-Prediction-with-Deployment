**Mengganti Nilai Outlier dengan nilai NaN**

```python
for column in numerical:
    data = df[column]
    
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    minimum = Q1 - (1.5 * IQR)
    maximum = Q3 + (1.5 * IQR)
 
    outliers = ((data < minimum) |(data > maximum))
    df[column].loc[outliers] = np.nan
    
df.isnull().sum()
```

Missing value pada atribut akan diisi dengan moving average, Moving Average adalah teknik untuk menghitung rata-rata data selama periode waktu tertentu.

**Mengisi Missing value dengan teknik Moving Average**

```python
# Kolom sisa
remnant = [cols for cols in df.columns if df[cols].isnull().sum() > 0 ]

# Mengisi missing values setiap kolom dengan moving average (rolling)
for cols in remnant:
    if df[cols].isnull().sum() != 0:
        while True:
            df[cols] = df[cols].fillna(df[cols].rolling(min_periods = 1, window = 7, center=True).mean())
            if df[cols].isnull().sum() == 0:
                    break
```

## 4. Handling Categorical Variable

Pada bagian ini Variabel kategorikal akan diubah nilainya menjadi nilai numerik, terdapat dua jenis Variabel Kategorikal, yaitu :
- Nominal Variabel : Data kategorikal yang hanya merepresentasikan labelnya saja tanpa ada penyediaan arti numerik, contohnya - Warna, Tempat, Tinggi, dll
- Ordinal Variabel : Data yang berbentuk non-numerik tetapi merepresentasikan arti numerik didalamnya (Scaling), Contoh : Nilai dalam bentuk Huruf (A > B > C > D > E)

Untuk Nominal Variabel akan diubah menjadi representasi numerik menggunakan data dummy, dan hanya akan menggunakan 1 variabel nominal

Untuk Ordinal Variabel akan diubah menjadi representasi numerik menggunakan LabelEncoder.

## 4. Algorithm Selection

```python
def train(algorithm, fiture, label) :
  
  model = algorithm
  model.fit(fiture, label)

  return model

def test(model, fiture_test) :

  ml = model
  predict = []

  for i in ml :
    predict.append(i.predict(fiture_test))
  
  return predict


def eval(pred, label_test, name) :
  
  val = pred
  accuracy = []
  recall = []
  precision = []

  for i in val : 
    predict = i
    accuracy.append(accuracy_score(label_test, predict))
    recall.append(recall_score(label_test, predict, pos_label=1))
    precision.append(precision_score(label_test, predict,  pos_label=1))

  dic = {
      'Model' : name,
      'accuracy' : accuracy,
      'recall' : recall,
      'precision' : precision
  }

  eval = pd.DataFrame(data = dic)
  
  return eval
```
